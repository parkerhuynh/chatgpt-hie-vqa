```python
import json
import pandas as pd
from processing import prep_ans
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('/home/ndhuynh/data/simpsonsvqa/v1_Annotation_Test_Private_110496_simpsons_vqa.json', 'r') as file:
       targets  = json.load(file)["annotations"]
```

```python
processed_targets = []
for ann in targets:
    ans_count = 0
    for judge in ann["judgements"].values():
        if judge["answer"] == 1:
            ans_count += 1
    if ans_count >= 2:
        ann["answer"] = prep_ans(ann["answer"])
        ann["original_answer"] = ann["answer"]
        processed_targets.append(ann)
    
```

```python
target = pd.DataFrame(processed_targets)
target = target[["id", "answer"]]
target = target.rename(columns = {"id": "question_id", "answer": "target"})
target
```

```python
prediction = pd.read_csv("predictions_vqa_simpsons.csv")
```

```python
prediction
```

```python
df = pd.merge(target, prediction, on="question_id")
```

```python
df
```

```python
print(f"Accuracy: {round(accuracy_score(df['target'],df['prediction'])*100,4)} %")
```
